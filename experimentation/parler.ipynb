```python
import torch
from parler_tts import ParlerTTSForConditionalGeneration
from transformers import AutoTokenizer
import soundfile as sf
from IPython.display import Audio

device = "cuda:0" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# Loading the TTS model and tokenizers
model = ParlerTTSForConditionalGeneration.from_pretrained("ai4bharat/indic-parler-tts").to(device)
tokenizer = AutoTokenizer.from_pretrained("ai4bharat/indic-parler-tts")
print("✅ Setup complete.")
```

```python
# The Hindi sentence you want to convert to speech
text_to_speak = "आर्टिफिशियल इंटेलिजेंस का भविष्य बहुत उज्ज्वल है। देश की प्रगति में शिक्षा और स्वास्थ्य सेवाओं का भी महत्वपूर्ण योगदान है।"

### IMP: A description of the voice you want to hear. You can be creative here!
voice_description = "A deep, warm voice speaking calmly and clearly, with a soft, soothing tone, recorded in a studio with no echo"
output_filename = "parler_tts_hindi_output.wav"
print(f"Synthesizing text: '{text_to_speak}'")
print(f"Using voice: '{voice_description}'")

# Tokenize the inputs-
prompt_input_ids = tokenizer(text_to_speak, return_tensors="pt").to(device)
description_input_ids = tokenizer(voice_description, return_tensors="pt").to(device)

# Generating the audio waveform
with torch.no_grad():
    generation = model.generate(
        input_ids=description_input_ids.input_ids,
        prompt_input_ids=prompt_input_ids.input_ids,
    )
audio_arr = generation.cpu().numpy().squeeze()



sf.write(output_filename, audio_arr, model.config.sampling_rate)
print(f"Audio saved to '{output_filename}'")

display(Audio(output_filename))
```

```python
# 1.8gb download here (one time)
import torch
from TTS.api import TTS
from IPython.display import Audio

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")


tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2").to(device)
print(f"✅ Coqui XTTS setup complete.")
```

```python
#debugging
dir(tts)
# print(dir(tts.synthesizer.tts_model.speaker_manager))
print(f"🗣️ Available speakers: {tts.synthesizer.tts_model.speaker_manager.name_to_id}")
```

```python
# give text here
text = "भारत एक विविध और विशाल देश है, जहां विभिन्न संस्कृतियों और भाषाओं का समावेश है। यहाँ के लोग अपनी मेहनत और संघर्ष से देश को आगे बढ़ाने में महत्वपूर्ण भूमिका निभाते हैं।"
output_filename = "hindi_xtts_output.wav"

tts.tts_to_file(text=text,                # VERY HELPFUL STEP BELOW, AS WE CAN CONTROL THE TONE AND SPEED OF SPEAKER!
                speaker="Tammie Ema",     # can change this to any speaker from the list above
                language="hi",
                file_path=output_filename)


display(Audio(output_filename))
print(f"Audio generated and saved as '{output_filename}'")
```

```python
import torch
from transformers import VitsModel, AutoTokenizer
from IPython.display import Audio

print("Setting up model and tokenizer...")

# model loading and tokenizer (ONE TIME)
model = VitsModel.from_pretrained("facebook/mms-tts-hin")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-hin")

print("✅ Setup complete. You can now run the next cell.")
```

```python
# --- Run this every time ---
text = "अब यह बहुत तेजी से काम करेगा क्योंकि मॉडल पहले से ही लोड हो चुका है।"

#text = "आज सप्ताह की नई शुरुआत होने के नाते, शहर के लोग अपने-अपने दफ्तरों की ओर जा रहे हैं, इसलिए प्रमुख सड़कों पर सामान्य से अधिक यातायात दिखाई दे रहा है।"

inputs = tokenizer(text, return_tensors="pt")
with torch.no_grad():
    output = model(**inputs).waveform

# for jupy notebook only
sampling_rate = model.config.sampling_rate
display(Audio(output.squeeze().numpy(), rate=sampling_rate))
```
